```python
# 라이브러리 그래프 꺽은선형 그래프
import matplotlib.pyplot as plt
import random
#data
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
# font
plt.rc('font', family ='Malgun Gothic')
# 제목
plt.title('오디션 프로그램 득표 현황')
plt.plot(singer, week1, label = '첫째주', color='hotpink')
plt.plot(singer, week2, label = '둘째주', color='royalblue')
# 축제목
plt.xlabel('가수')
plt.ylabel('점수')
# 범례
plt.legend()
plt.show()
```

```python
# 라이브러리 세로막대 그래프
import matplotlib.pyplot as plt
import random
#data
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
# font
plt.rc('font', family ='Malgun Gothic')
# 제목
plt.title('오디션 프로그램 득표 현황')
plt.bar(range(1,14,3), week1, label = '첫째주', color='hotpink')
plt.bar(range(2,15,3), week2, label = '둘째주', color='royalblue')
plt.xticks(range(1,14,3), singer)
# 범례
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
# font
plt.rc('font', family ='Malgun Gothic')

employee = ['a', 'b', 'c']
incen2018 = [50, 91, 37]
incen2019 = [18, 67, 93]

plt.title('직원 성과급 현황(2018~2019)')
plt.bar(range(1, 8, 3), incen2018, label='2018년')
plt.bar(range(2, 9, 3), incen2019, label='2019년')
plt.xticks(range(1, 8, 3), employee)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]

plt.rc('font', family='Malgun Gothic')
plt.title('주중/주말 자동차 통행량 평일/주말 비교')
plt.plot(range(8), weekday, color='crimson', label = '주중 통행량')
plt.plot(range(8), weekend, color ='royalblue', label = '주말 통행량')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]

for i in range(len(weekday)) :
    weekday[i] = -weekday[i]

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('주중/주말 자동차 통행량 평일/주말 비교')
plt.barh(range(8), weekday, color='crimson', label = '주중 통행량')
plt.barh(range(8), weekend, color ='royalblue', label = '주말 통행량')
plt.legend()
plt.show()
```
