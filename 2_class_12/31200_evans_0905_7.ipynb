```python
# 산점도 그래프 랜덤
# 라이브러리 그래프
import matplotlib.pyplot as plt
import random
#data
height, weight, score =[], [], []

for i in range(100) :
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
# font
plt.rc('font', family ='Malgun Gothic')
# 제목
plt.title('키와 몸무게의 상관관계')
plt.scatter(height, weight, s = score, c = score, cmap='rainbow', alpha=0.7)
plt.colorbar(label='체육 실기점수')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

```python
# 라이브러리 그래프 꺽은선형 그래프
import matplotlib.pyplot as plt
import random
#data
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
# font
plt.rc('font', family ='Malgun Gothic')
# 제목
plt.title('오디션 프로그램 득표 현황')
plt.plot(singer, week1, label = '첫째주', color='hotpink')
plt.plot(singer, week2, label = '둘째주', color='royalblue')
# 축제목
plt.xlabel('가수')
plt.ylabel('점수')
# 범례
plt.legend()
plt.show()
```

```python
# 라이브러리 가로막대 그래프
import matplotlib.pyplot as plt
import random
#data
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
for i in range(len(week1)) :
    week1[i] = -week1[i]
# font
plt.rc('font', family ='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 표시
# 제목
plt.title('오디션 프로그램 득표 현황')
plt.barh(singer, week1, label = '첫째주', color='hotpink')
plt.barh(singer, week2, label = '둘째주', color='royalblue')
# 축제목
plt.xlabel('점수')
plt.ylabel('가수')
# 범례
plt.legend()
plt.show()
```

```python
# 산점도 그래프
import matplotlib.pyplot as plt
import random
# font
plt.rc('font', family ='Malgun Gothic')

# 데이터 만들기
height1, weight1, height2, weight2 =[], [], [], []
for i in range(100) :
    height1.append(random.randint(140,180))
    weight1.append(random.randint(40,80))
    height2.append(random.randint(160,200))
    weight2.append(random.randint(50,100))

# 제목
plt.title('키와 몸무게의 상관관계')

# 산점도 그래프 2개 그리기
plt.scatter(height1, weight1, color = 'crimson', alpha = 0.7, label='그룹1')
plt.scatter(height2, weight2, color = 'indigo', alpha = 0.7, label='그룹2')
plt.legend() #범례
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

```python
# 산점도 그래프
import matplotlib.pyplot as plt
import random
# font
plt.rc('font', family ='Malgun Gothic')

# 데이터 만들기 2그룹
height1, weight1, score1, height2, weight2, score2 = [], [], [], [], [], []

for i in range(100):
    height1.append(random.randint(140, 180))
    weight1.append(random.randint(40, 80))
    score1.append(random.randint(0, 200))
    height2.append(random.randint(160, 200))
    weight2.append(random.randint(50, 100))
    score2.append(random.randint(0, 200))

plt.scatter(height1, weight1, s=score1, color='crimson', alpha=0.7, label='그룹1')
plt.scatter(height2, weight2, s=score2, color='indigo', alpha=0.7, label='그룹2')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]

plt.rc('font', family='Malgun Gothic')
plt.title('주중/주말 자동차 통행량 평일/주말 비교')
plt.plot(range(8), weekday, color='crimson', label = '주중 통행량')
plt.plot(range(8), weekend, color ='royalblue', label = '주말 통행량')
plt.legend()
plt.show()
```
