```python
#   
#  
import matplotlib.pyplot as plt
import random
#data
height, weight, score =[], [], []

for i in range(100) :
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
# font
plt.rc('font', family ='Malgun Gothic')
# 
plt.title('  ')
plt.scatter(height, weight, s = score, c = score, cmap='rainbow', alpha=0.7)
plt.colorbar(label=' ')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
#   
import matplotlib.pyplot as plt
import random
#data
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
for i in range(len(week1)) :
    week1[i] = -week1[i]
# font
plt.rc('font', family ='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False #   
# 
plt.title('   ')
plt.barh(singer, week1, label = '', color='hotpink')
plt.barh(singer, week2, label = '', color='royalblue')
# 
plt.xlabel('')
plt.ylabel('')
# 
plt.legend()
plt.show()
```

```python
#  
import matplotlib.pyplot as plt
import random
# font
plt.rc('font', family ='Malgun Gothic')

#   2
height1, weight1, score1, height2, weight2, score2 = [], [], [], [], [], []

for i in range(100):
    height1.append(random.randint(140, 180))
    weight1.append(random.randint(40, 80))
    score1.append(random.randint(0, 200))
    height2.append(random.randint(160, 200))
    weight2.append(random.randint(50, 100))
    score2.append(random.randint(0, 200))

plt.scatter(height1, weight1, s=score1, color='crimson', alpha=0.7, label='1')
plt.scatter(height2, weight2, s=score2, color='indigo', alpha=0.7, label='2')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]

for i in range(len(weekday)) :
    weekday[i] = -weekday[i]

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('/   / ')
plt.barh(range(8), weekday, color='crimson', label = ' ')
plt.barh(range(8), weekend, color ='royalblue', label = ' ')
plt.legend()
plt.show()
```
