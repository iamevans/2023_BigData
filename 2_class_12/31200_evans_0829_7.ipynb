```python
#   
#  
import matplotlib.pyplot as plt
import random
#data
height, weight, score =[], [], []

for i in range(100) :
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
# font
plt.rc('font', family ='Malgun Gothic')
# 
plt.title('  ')
plt.scatter(height, weight, s = score, c = score, cmap='rainbow', alpha=0.7)
plt.colorbar(label=' ')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
#    
import matplotlib.pyplot as plt
import random
#data
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
# font
plt.rc('font', family ='Malgun Gothic')
# 
plt.title('   ')
plt.plot(singer, week1, label = '', color='hotpink')
plt.plot(singer, week2, label = '', color='royalblue')
# 
plt.xlabel('')
plt.ylabel('')
# 
plt.legend()
plt.show()
```

```python
#    
import matplotlib.pyplot as plt
import random
#data
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
for i in range(len(week1)) :
    week1[i] = -week1[i]

# font
plt.rc('font', family ='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False #   
# 
plt.title('   ')
plt.barh(singer, week1, label = '', color='hotpink')
plt.barh(singer, week2, label = '', color='royalblue')
# 
plt.xlabel('')
plt.ylabel('')
# 
plt.legend()
plt.show()
```
