```python
# 라이브러리 불러오기 import
# 그래픽 라이브러리 matplotlib.pyplot

import matplotlib.pyplot as plt

# x축,y축 데이터
mon = ['mar','apr','may','jun','jul']
sales = [1, 5, 7, 3, 7]

# 한글 폰트 적용
plt.rc('font', family='Malgun Gothic')

# 그래프 제목
plt.title('월별 판매 실적')

# 꺽은선형 그래프 그리기 x,y,color, 범례 label
plt.plot(mon, sales, color='red', label = '단위:개수')
#범례위치
plt.legend() #왼쪽위

plt.show()
```

```python
# 바차트 그래프 bar

plt.title('월별 판매 실적 세로형 막대그래프')
plt.bar(mon, sales, color='green', label='단위:개수')
plt.legend()
plt.show()
```

```python
# 바차트 그래프 bar

plt.title('월별 판매 실적 세로형 막대그래프')

# range(시작숫자, 종료숫자, 간격)
plt.bar(range(2,12,2), sales, color='hotpink', label='단위:개수')
plt.legend()
plt.show()
```
