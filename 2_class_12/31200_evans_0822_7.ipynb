````python
# 라이브러리 불러오기 import
# 그래픽 라이브러리 matplotlib.pyplot
import matplotlib.pyplot as plt

# x축,y축 데이터
mon = ['mar','apr','may','jun','jul']
sales = [1, 5, 7, 3, 7]

# 한글 폰트 적용
plt.rc('font', family='Malgun Gothic')

# 그래프 제목
plt.title('월별 판매 실적')

# 꺽은선형 그래프 그리기 x,y,color, 범례 label
plt.plot(mon, sales, color='red', label = '단위:개수')
#범례위치
plt.legend() #왼쪽위
plt.show()
````

````python
# 바차트 그래프 bar

plt.title('월별 판매 실적 세로형 막대그래프')
plt.bar(mon, sales, color='green', label='단위:개수')
plt.legend()
plt.show()
````

````python
# 바차트 그래프 bar

plt.title('월별 판매 실적 세로형 막대그래프')

# range(시작숫자, 종료숫자, 간격)
plt.bar(range(2,12,2), sales, color='hotpink', label='단위:개수')
plt.legend()
plt.show()
````

````python
# 바차트 그래프 barh

plt.title('월별 판매 실적 가로형 막대그래프')
plt.barh(mon, sales, color='hotpink', label='단위:개수')
plt.legend()
plt.show()
````

````python
# 히스토그램 빈돗수

print(sales)
````

````python
plt.rc('font', family='Malgun Gothic')
plt.title('판매 실적 히스토그램')
# 히스토그램 차트 hist
plt.hist(sales)
plt.show()
````

### plt.pie() : 원그래프
```
우리학교 100명의 혈액형을 조사한 결과가 다음과 같다.

A형 = 28명
B형 = 20명
O형 = 40명
AB형 = 12명
```

````python
b_type = [28,20,40,12]

# 원형차트 pie
plt.pie(b_type)
plt.show()
````

#### 원그래프 각 항목에 라벨 붙이기
```
plt.pie(리스트, labels = 라벨리스트)
```

````python
# 라이브러리 불러오기 import
# 그래픽 라이브러리 matplotlib.pyplot
import matplotlib.pyplot as plt

# 데이터
b_type = [28,20,40,12]
b_name = ['A형', 'B형', 'O형', 'AB형']

# 한글 폰트 적용
plt.rc('font', family='Malgun Gothic')
# 원형차트 제목
plt.title('혈액형 비율')
# 원형차트 그리기 퍼센트 추가 autopct='%1.1f%%'
plt.pie(b_type, labels = b_name, autopct='%1.1f%%')
plt.show()
````

### plt.scatter() : 산점도=점도표

> 국어를 잘하는 친구가 영어도 잘할까?

```
국어점수와 영어점수의 상관관계 알아보기

kor =[80, 20, 50, 20, 10, 50, 60, 30, 60]
eng =[90, 40, 60, 40, 10, 30, 50, 70, 90]
```

### 그래프 꾸미기 : x축 이름, y축 이름 붙이기
```
plt.xlabel('x축 이름')
plt.ylabel('y축 이름')
```

````python
# 라이브러리 불러오기 import
# 그래픽 라이브러리 matplotlib.pyplot
import matplotlib.pyplot as plt
# 한글 폰트 적용
plt.rc('font', family='Malgun Gothic')

#데이터
kor =[80, 20, 50, 20, 10, 50, 60, 30, 60]
eng =[90, 40, 60, 40, 10, 30, 50, 70, 90]

#제목
plt.title('국어점수와 영어점수와의 상관관계')
# 그래프 산점도
plt.scatter(kor, eng, c='indigo')
# 축제목
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.show()
````
