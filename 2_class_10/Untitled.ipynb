```python
#import pandas as pd
import csv 

# 파일 오픈
f = open('card.csv', 'rt', encoding='UTF8')         
data = csv.reader(f)         # 파일 읽기 reader
next(data)                   # 파일 헤더 건너뛰기 
data = list(data)            # 파일 리스트로 바꾸기
```

```python
# 카드이용일자 모두 출력하기
for row in data : 
    print(row[0])
```

```python
# 카드 사용금액 모두 출력하기
for row in data : 
    print(row[-3])
```

```python
# 카드 사용처와 금액 모두 출력하기
for row in data : 
    print(row[5],'에서',row[6],'원 사용')
```

```python
# 카드 사용처와 금액 모두 출력하기
for row in data : 
    store = row[5]
    payment = row[6]
    print(store,'에서',payment,'원 사용')
```

```python
spend = 0

for row in data : 
    spend += int(row[-3])

print(spend)
```

```python
spend = 0

for row in data : 
    if row[-1] == '전표매입' : 
        spend += int(row[-3])

print(spend)
```

```python
s_mon = [0, 0, 0]

for row in data : 
    if row[-1] == '전표매입' : 
        mon = row[0].split('-')[1]   
        idx = mon - 10                
        payment = int(row[6])
        s_mon[idx] += payment

print(s_mon)
```

```python
s_mon = [0, 0, 0]

for row in data : 
    if row[-1] == '전표매입' : 
        mon = int(row[0].split('-')[1]) 
        idx = mon - 10                
        payment = int(row[6])
        s_mon[idx] += payment

print(s_mon)
```

```python
# 초간단ver
s_mon = [0, 0, 0]

for row in data : 
    if row[-1] == '전표매입' : 
        s_mon[int(row[0].split('-')[1]) - 10 ] += int(row[6])

print(s_mon)
```

```python
import matplotlib.pyplot as plt

# 그래프 해상도
plt.figure(dpi=100)
# 그래프 폰트
plt.rc('font',family ='NanumGothic')
# 그래프 제목
plt.title('10-12월 지출현황')
# 그래프 유형 bar
plt.bar(['10월','11월','12월'], s_mon, color = 'royalblue')
plt.show()
```

```python
taxi = [0, 0, 0]

for row in data : 
    if row[-1] == '전표매입' and '택시' in row[5] : 
        mon = int(row[0].split('-')[1])
        payment = int(row[-3])
        idx = mon-10
        taxi[idx] += payment

print(taxi)
```

```python
# 초간단 ver
taxi = [0,0,0]

for row in data : 
    if row[-1] == '전표매입' and '택시' in row[5] : 
        taxi[int(row[0].split('-')[1])-10] += int(row[-3])

print(taxi)
```

```python
plt.figure(dpi=100)
plt.title('10-12월 택시비 지출 현황')
plt.plot(['10월','11월','12월'],taxi, color = 'crimson', label ='택시 지출액')
plt.legend()
plt.show()
```

```python
deli = [0,0,0]

for row in data : 
    if row[-1] == '전표매입' and row[5] == '(주)우아한형제들': 
        mon = int(row[0].split('-')[1])
        idx = mon-10
        payment = int(row[-3])        
        deli[idx] += payment

print(deli)
```

```python
plt.figure(dpi=100)

plt.title('10-12월 배달음식 지출 현황')
plt.plot(['10월','11월','12월'], deli, color = 'indigo', label ='배달음식 지출액')
plt.legend()
plt.show()
```

```python
plt.figure(dpi=100)

plt.title('10-12월 택시/배달음식 지출 현황')

plt.plot(['10월', '11월','12월'], taxi, color = 'crimson', label = '택시 지출액')
plt.plot(['10월','11월','12월'], deli, color = 'indigo', label='배달음식 지출액')

plt.legend()
plt.show()
```
