````python
# 꺽은선형 그래프 그리기
# plot

# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')

# 그래프 데이터 설정
mon = ['mar','apr','may','jun','jul'] #x축
sales = [1, 5, 7, 3, 7] #y축

# 그래프 제목
plt.title('월별 판매 실적 : 꺽은선형 그래프')
# 그래프 그리기  x축값, y축값, 색상, 범례
plt.plot(mon, sales, color='red', label='단위:개수')
# 범례위치 지정
plt.legend() #default 왼쪽위
plt.show()
````

````python
# 막대그래프 그리기
# bar

# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')

# 그래프 데이터 설정
mon = ['mar','apr','may','jun','jul'] #x축
sales = [1, 5, 7, 3, 7] #y축

# 제목
plt.title('월별 판매 실적 : 세로형 막대 그래프')

# 막대그리프 그리기
plt.bar(mon, sales, color='green', label='단위:개수')
plt.legend()
plt.show()
````

### range(시작숫자, 종료숫자, 간격)

시작숫자부터 종료숫자까지 간격을 두어 숫자를 선택함 <br/>
```
range(2, 11, 2)
```
2부터 11까지 간격 2를 두고 숫자 선택
-> 2, 4, 6, 8, 10

````python
plt.title('월별 판매 실적 : x축 데이터 범위 설정')
plt.bar(range(2,12,2), sales, color='blue')
plt.show()
````

````python
# plt.barh() : 막대그래프 - 가로형
plt.title('월별 판매 실적 가로형 막대 그래프')
plt.barh(range(3,8), sales, color = 'hotpink')
plt.show()
````

### plt.pie() : 원그래프
```
우리학교 100명의 혈액형을 조사한 결과가 다음과 같다.
A형 = 28명
B형 = 20명
O형 = 40명
AB형 = 12명
```

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')

# data
b_type = [28, 30, 40, 12]

# 원그래프 pie
plt.pie(b_type)
plt.show()
````

#### 원그래프 각 항목에 라벨 붙이기
```
plt.pie(리스트, labels = 라벨리스트)
```

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')

# data labels
b_type = [28, 30, 40, 12]
b_name = ['A형', 'B형', 'O형', 'AB형']

# 제목
plt.title('혈액형 비율 : 원형 그래프')
# 원형 그래프 autopct='%1.1f%%'
plt.pie(b_type, labels = b_name, autopct='%1.1f%%')
plt.show()
````

### plt.scatter() : 산점도

> 국어를 잘하는 친구가 영어도 잘할까?
```
국어점수와 영어점수의 상관관계 알아보기

kor =[80, 20, 50, 20, 10, 50, 60, 30, 60]
eng =[90, 40, 60, 40, 10, 30, 50, 70, 90]
```

````python
# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')

# data
kor =[80, 20, 50, 20, 10, 50, 60, 30, 60]
eng =[90, 40, 60, 40, 10, 30, 50, 70, 90]
# 제목
plt.title('국어점수와 영어점수와의 상관관계')

# 산점도 그래프 그리기
plt.scatter(kor, eng, c='indigo')
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.show()
````

````python
# 그래프 라이브러리 가져오기
import matplotlib.pyplot as plt
# 랜덤함수 가져오기
import random
# data
height = []
weight = []
for i in range(100) :
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
# font
plt.rc('font', family='Malgun Gothic')
# title
plt.title('키와 몸무게의 상관관계')
# 산점도 그래프 scatter
plt.scatter(height, weight, c= height, cmap='RdPu')
plt.colorbar(label='키')
# 축제목
plt.xlabel('키')
plt.ylabel('몸무게')
# 그래프를 보여줘
plt.show()
````

````python
# 그래프 라이브러리 가져오기
import matplotlib.pyplot as plt
# 랜덤함수 가져오기
import random
# data
height, weight, score = [],[],[]
for i in range(100) :
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
                 
# font
plt.rc('font', family='Malgun Gothic')
# title
plt.title('키와 몸무게의 상관관계')
# 산점도 그래프
plt.scatter(height,weight,s = score, c = score, cmap='rainbow', alpha=0.7)
plt.xlabel('키')
plt.ylabel('몸무게')
plt.colorbar(label='체육실기점수')
plt.show()
````

````python
# 상자 수염 그림 그리기
# 그래프 라이브러리 가져오기
import matplotlib.pyplot as plt

# data 
kor = [90, 20, 50, 20, 10, 50, 60, 30, 60]

# 상자 수염 그림 그리기 boxplot
plt.boxplot(kor, labels=['국어점수'])
plt.show()
````

````python
#꺾은선 그래프 겹쳐 그리기
import matplotlib.pyplot as plt
# font
plt.rc('font', family='Malgun Gothic')
#data
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
#꺽은선형 그래프 그리기
# title
plt.title('오디션 프로그램 득표 현황')

plt.plot(singer, week1, label = '첫째주', color='hotpink')
plt.plot(singer, week2, label = '둘째주', color='royalblue')
plt.xlabel('가수')
plt.ylabel('점수')
plt.legend()
plt.show()
````

````python
#가로 막대 그래프 겹쳐 그리기 barh
import matplotlib.pyplot as plt
# font
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 표시
#data
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
for i in range(len(week1)) :
    week1[i] = -week1[i]

# title
plt.title('오디션 프로그램 득표 현황')
# 가로막대 그래프
plt.barh(singer, week1, label = '첫째주', color='hotpink')
plt.barh(singer, week2, label = '둘째주', color='royalblue')
plt.xlabel('점수')
plt.ylabel('가수')
plt.legend()
plt.show()
````

````python
#세로 막대 그래프 겹쳐 그리기 bar
import matplotlib.pyplot as plt
# font
plt.rc('font', family='Malgun Gothic')
#data
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

# title
plt.title('오디션 프로그램 득표 현황')
#세로 막대 그래프
plt.bar(range(1,14,3), week1, label = '첫째주', color='hotpink')
plt.bar(range(2,15,3), week2, label = '둘째주', color='royalblue')
plt.xticks(range(1,14,3),singer)

plt.legend()
plt.show()
````
