````python
# 꺽은선형 그래프 그리기
# plot

# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')


# 그래프 데이터 설정
mon = ['mar','apr','may','jun','jul'] #x축
sales = [1, 5, 7, 3, 7] #y축

# 그래프 제목
plt.title('월별 판매 실적')
# 그래프 그리기  x축값, y축값, 색상, 범례
plt.plot(mon, sales, color='red', label='단위:개수')
# 범례위치 지정
plt.legend() #default 왼쪽위
plt.show()
````

````python
# 막대그래프 그리기
# bar

# 그래픽 라이브러리 불러오기
import matplotlib.pyplot as plt
# 한글 폰트 사용하기
plt.rc('font', family ='Malgun Gothic')


# 그래프 데이터 설정
mon = ['mar','apr','may','jun','jul'] #x축
sales = [1, 5, 7, 3, 7] #y축

# 제목
plt.title('월별 판매 실적')

# 막대그리프 그리기
plt.bar(mon, sales, color='green', label='단위:개수')
plt.legend()
plt.show()
````

### range(시작숫자, 종료숫자, 간격)

시작숫자부터 종료숫자까지 간격을 두어 숫자를 선택함 <br/>
```
range(2, 11, 2)
```
2부터 11까지 간격 2를 두고 숫자 선택
-> 2, 4, 6, 8, 10

````python
plt.title('월별 판매 실적')
plt.bar(range(2,12,2), sales, color='blue')
plt.show()
````

````python
# plt.barh() : 막대그래프 - 가로형
plt.title('월별 판매 실적 가로형 막대 그래프')
plt.barh(range(3,8), sales, color = 'hotpink')
plt.show()
````
