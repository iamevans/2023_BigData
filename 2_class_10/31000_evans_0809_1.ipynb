매출그래프 그리기 
```
3-7월 매출

3월 = 1
4월 = 5
5월 = 7
6월 = 3
7월 = 7
```

````python
# 꺽은선형 그래프 그리기

import matplotlib.pyplot as plt

# 그래프를 그리는 명령어는 plot()
plt.plot([1,5,7,3,7])
# 그래프를 보여줘 명령어 show()
plt.show()
````

````python
sales = [1,5,7,3,7]

# list name 적용
plt.plot(sales)
plt.show()
````

### x축 추가하기

```
plt.plot(x축, y축)
plt.show()
```

````python
# x축 추가
mon = ['mar','apr','may','jun','jul']
sales = [1, 5, 7, 3, 7]

# x축, y축 적용한 그래프 그리기
plt.plot(mon, sales)
plt.show()
````

### 그래프 제목 추가하기
```
plt.title('제목')
plt.plot(x축, y축)
plt.show()
```

````python
# 그래프의 제목 출력
plt.title('monthly sales')
# 그래프 x축 y축 데이터 적용
plt.plot(mon, sales)
# 그래프를 보여줘
plt.show()
````
