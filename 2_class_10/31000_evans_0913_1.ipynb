```python
# 세로막대 그래프 2개 같이 그리기
# 라이브러리 불러오기
import matplotlib.pyplot as plt
# font
plt.rc('font', family='Malgun Gothic')
# 데이터 만들기
employee = ['a', 'b', 'c'] 
incen2018 = [50, 91, 37] 
incen2019 = [18, 67, 93]
# 그래프 제목
plt.title('직원 성과급 현황(2018~2019)') 
# 세로막대 그래프 그리기
plt.bar(range(1, 8, 3), incen2018, label='2018년') 
plt.bar(range(2, 9, 3), incen2019, label='2019년') 
plt.xticks(range(1, 8, 3), employee)
# 범례
plt.legend() 
plt.show()
```

```python
# 꺽은선형 그래프 2개 같이 그리기
# 라이브러리 불러오기
import matplotlib.pyplot as plt
# font
plt.rc('font', family='Malgun Gothic')
# 데이터 만들기
weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]
# 그래프 제목
plt.title('주중/주말 자동차 통행량 평일/주말 비교')
# 꺽은선 그래프 그리기
plt.plot(range(8), weekday, color='crimson', label = '주중 통행량')
plt.plot(range(8), weekend, color ='royalblue', label = '주말 통행량')
#범례
plt.legend()
plt.show()
```

```python
# 가로막대 그래프 2개 겹치기
# 라이브러리 불러오기
import matplotlib.pyplot as plt
# font
plt.rc('font', family='Malgun Gothic')
# 데이터 만들기
weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]
# weekday 마이너스 표시하기
for i in range(len(weekday)) : 
    weekday[i] = -weekday[i]
# 마이너스 데이터 표시    
plt.rcParams['axes.unicode_minus'] = False
# 그래프 제목 표시
plt.title('주중/주말 자동차 통행량 평일/주말 비교')
# 가로막대 그래프 그리기
plt.barh(range(8), weekday, color='crimson', label = '주중 통행량')
plt.barh(range(8), weekend, color ='royalblue', label = '주말 통행량')
#범례
plt.legend()
plt.show()
```

```python
# 상자수염 그래프 2개 겹치기
# 라이브러리 불러오기
import matplotlib.pyplot as plt
# font
plt.rc('font', family='Malgun Gothic')
# 데이터 만들기
weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]

# 그래프제목
plt.title('주중/주말 자동차 통행량 평일/주말 비교')
# 상자수염 그래프 그리기
plt.boxplot([weekday, weekend], labels=['주중 통행량','주말 통행량'])
plt.show()
```
