```python
# 가로막대 그래프 2개 겹치기
# 라이브러리 불러오기 matplotlib.pyplot
import matplotlib.pyplot as plt

# 한글폰트
plt.rc('font', family='Malgun Gothic')

# data
weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]
# 주중데이터의 표시를 마이너스를 붙여 봅니다.
for i in range(len(weekday)) :
    weekday[i] = -weekday[i]
    
# 그래프에서 마이너스값을 표시 세팅
plt.rcParams['axes.unicode_minus'] = False

# title
plt.title('주중/주말 자동차 통행량 평일/주말 비교')
# 가로막대 그래프 그리기
plt.barh(range(8), weekday, color='crimson', label = '주중 통행량')
plt.barh(range(8), weekend, color='royalblue', label = '주말 통행량')

# 범례
plt.legend()
plt.show()

#print(weekday)
#print(weekend)
```

```python
# 상자수염 그래프 2개 겹치기
# 라이브러리 불러오기
import matplotlib.pyplot as plt

# 한글폰트
plt.rc('font', family ='Malgun Gothic')

# data
weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]

# 제목 주중/주말 자동차 통행량 평일/주말 비교
plt.title('주중/주말 자동차 통행량 평일/주말 비교')

# 상자수염 그래프 그리기 boxplot
plt.boxplot([weekday, weekend], labels=['주중 통행량', '주말 통행향'])
plt.show()
```

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt

# 한글 폰트
plt.rc('font', family='Malgun Gothic')

# data
weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]

# title
plt.title('주중/주말 자동차 통행량 평일/주말 비교')

# 꺽은선형 그래프 그리기 plot
plt.plot(range(8), weekday, color='crimson', label ='주중 통행량')
plt.plot(range(8), weekend, color='royalblue', label ='주말 통행량')

# 범례
plt.legend()
plt.show()
```
