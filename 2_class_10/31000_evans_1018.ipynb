```python
#   2 
#   matplotlib.pyplot
import matplotlib.pyplot as plt

# 
plt.rc('font', family='Malgun Gothic')

# data
weekday = [3, 2, 16, 19, 9, 12, 23, 16]
weekend = [1, 7, 13, 15, 19, 10, 19, 12]
#     .
for i in range(len(weekday)) :
    weekday[i] = -weekday[i]
    
#    
plt.rcParams['axes.unicode_minus'] = False

# title
plt.title('/   / ')
#   
plt.barh(range(8), weekday, color='crimson', label = ' ')
plt.barh(range(8), weekend, color='royalblue', label = ' ')

# 
plt.legend()
plt.show()

#print(weekday)
#print(weekend)
```
