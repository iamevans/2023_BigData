매출그래프 그리기 

```
3-7월 매출

3월 = 1
4월 = 5
5월 = 7
6월 = 3
7월 = 7
```

````python
# 꺽은선형 그래프 그리기
# 필요한 라이브러리 가져오기 matplotlib.pyplot
import matplotlib.pyplot as plt

plt.plot([1,5,7,3,7])
plt.show()
````

````python
sales = [1,5,7,3,7]

# y축 값을 이름 부여

plt.plot(sales)
plt.show()
````

### x축 추가하기

```
plt.plot(x축, y축)
plt.show()
```

````python
# x축 추가
mon = ['mar','apr','may','jun','jul']
sales = [1, 5, 7, 3, 7]

# X축, y축
plt.plot(mon, sales)
plt.show()
````

### 그래프 제목 추가하기
```
plt.title('제목')
plt.plot(x축, y축)
plt.show()
```

````python
plt.title('월별 monthly sales')
plt.plot(mon, sales)
plt.show()
````

````python
import matplotlib
import matplotlib.font_manager as fm
fm.get_fontconfig_fonts()
font_location = './NanumGothic.ttf' # 폰트 파일 이름, 디렉토리 주의
matplotlib.rc('font', family=NanumGothic.ttf)
````

````python
import matplotlib.pyplot as plt

mon = ['mar','apr','may','jun','jul']
sales = [1,5,7,3,7]

# 한글 폰트 설정

plt.title('monthy sales')
plt.plot(mon, sales)
plt.show()
````

### 그래프 색상 변경
```
# 그래프 그리는 명령어에 color 옵션 추가
plt.plot(x축, y축, color = 'r')
```

````python
plt.title('monthy sales')
# 그래프 색상 변경
plt.plot(mon, sales, color ='r')
plt.show()
````
