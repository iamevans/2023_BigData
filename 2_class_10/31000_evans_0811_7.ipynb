매출그래프 그리기

```
3-7월 매출

```

````python
# 라이브러리 libray matplotlib.pyplot
import matplotlib.pyplot as plt

plt.plot([1,5,7,3,7])
plt.show()
````

````python
sales = [1,5,7,3,7]

plt.plot(sales)
plt.show()
````

### x축 추가하기
```
plt.plot(x축, y축)
plt.show()
```

````python
# x축 추가
mon = ['mar','apr','may','jun','jul']
sales = [1, 5, 7, 3, 7]

# X축, y축
plt.plot(mon, sales)
plt.show()
````

### 그래프 제목 추가하기
```
plt.title('제목')
plt.plot(x축, y축)
plt.show()
```

````python
#한글폰트 가져오기
plt.rc('font', family = "Malgun Gothic")


plt.title('월별 판매 실적')
plt.plot(mon, sales, color = "green")
plt.show()
````

### 범례 나타내기 색상
```
plt.plot(x축, y축, color = 'red', label = '범례 내용')
plt.legend()      # 범례를 화면에 나타내기
plt.show()
```

````python
#한글폰트 가져오기
plt.rc('font', family = "Malgun Gothic")
# 제목
plt.title('월별 판매 실적')
# 범례 label
plt.plot(mon, sales, color = 'blue', label = '단위:개수')
# 범례를 화면에 나타내는 legend()
# 범례위치 loc='upper right', 'lower left', 'center left', 'center' 
plt.legend(loc='lower left')
plt.show()
````
